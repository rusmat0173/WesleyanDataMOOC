PartyIDN values less than 1 or greater than 9 are Other party, Other Answer, Don't Knows, Refusals, etc.  So wish to filter these out. <br> Do same for FraudSnt

```python
filter01= dfbsa.loc[(dfbsa['PartyIDN'] == 1) | (dfbsa['PartyIDN'] == 2) | (dfbsa['PartyIDN'] == 3) | (dfbsa['PartyIDN'] == 4) | (dfbsa['PartyIDN'] == 5) |  (dfbsa['PartyIDN'] == 6) |  (dfbsa['PartyIDN'] == 7) |  (dfbsa['PartyIDN'] == 8) | (dfbsa['PartyIDN'] == 9)]
filter02 = filter01.loc[(filter01['FraudSnt'] == 1) | (filter01['FraudSnt'] == 2) | (filter01['FraudSnt'] == 3) | (filter01['FraudSnt'] == 4) | (filter01['FraudSnt'] == 5) |  (filter01['FraudSnt'] == 6)]
```

```python
filtered_df = filter02.filter(['PartyIDN','FraudSnt'])
filtered_df.head()
```

```python
print("length of filtered df: ", len(filtered_df))
# next categoric from numeric 
filtered_df = filtered_df.astype('category')
```

As per lecture, progress to chi-square analysis.<br>
### Null hypothesis
If there were no party affiliation-bias, the distribution of responses to views about media reporting of social security fraud would be very similar whatever the party-political affiliation.<br>
In other words the NULL HYPOTHESIS is that there is no relationship between views about media reporting of social security fraud and party-political affiliation.

```python
# contingency table of observed counts
ct1=pd.crosstab(filtered_df['FraudSnt'], filtered_df['PartyIDN'])
ct1
```

```python
# column percentages
colsum=ct1.sum(axis=0)
colpct=ct1/colsum
colpct
```

The FraudSnt categories are as follows: "Is the media critical or sympathetic in the way it portrays people people who engage in benefit fraud?"<br>
1 = Very critical, 2 = Somewhat critical, 3 = Neither critical nor sympathetic, 4 = Somewhat sympathetic, 5 = Very sympathetic, 6 = Mixed 

```python
# chi-square
print ('chi-square value, p value, expected counts')
cs1= scipy.stats.chi2_contingency(ct1)
cs1
```

### The result
The p-value above is 0.05927..., which is larger than 0.05. We say the result is not significant enough to reject the null hypothesis. (If we did accept the alternative hypothesis we would be wrong more than 5% of the time.)
Actually, 0.05927 is not that much greater than 0.05, the usual p-value cut-off for significance.<br>
So what can we say about such 'closeness'? The following article urges the researcher to stay strict about 0.05 as significance, and not to 'move the goal posts':<br>
http://blog.minitab.com/blog/understanding-statistics/what-can-you-say-when-your-p-value-is-greater-than-005

### Interpretation
The best place to look is the column percentages table, above. We are comparing columns percentages across each explanatory category. (Here, political party affiliation.) Since the null hypothesis could not be rejected, we can say that the difference in column distributions across all party affiliations are not different enough



### Post hoc tests
What do the 'post hoc tests' test for here?  They would test by 'pairwise' political party affiliation whether the views about media reporting of social security fraud are significantly different. <br>
It is clear from inspection of the pandas cross tab table (above) that political parties 8 and 9 had a very small sample size, so I will not do such comparisons that include those parties. (Already, with parties 1-7 there are 6+5+4+3+2+1=21 pairwise tests to do!)<br>
Aim is to automate the process of 21 pairwise comparisons using Python code

```python
# create list of party numbers
party_list = [x+1 for x in range(7)]
# create a list of dictionaries that will go into the recode variable that Week 2's code uses for pairwise comparisons
my_list = []
for i in range(len(party_list)):
    count = 0
    while i+1+count <=(len(party_list)):
        temp_var =({i+1: i+1, i+1+count: i+1+count})
        count+=1
        if len(temp_var) == 2:
            my_list.append(temp_var)
print("list of recode dictionaries here:", my_list)
# now use the recode dictionaries in my_list, and just use Week 2's code for pairwise comparisons
for item in my_list:
    recode2 = item
    filtered_df['PartyIDNv2']= filtered_df['PartyIDN'].map(recode2)

    # contingency table of observed counts
    ct2=pd.crosstab(filtered_df['FraudSnt'], filtered_df['PartyIDNv2'])
    print (ct2)

    # column percentages
    colsum=ct2.sum(axis=0)
    colpct=ct2/colsum
    print(colpct)

    print ('chi-square value, p value, expected counts')
    cs2= scipy.stats.chi2_contingency(ct2)
    print (cs2)
```
